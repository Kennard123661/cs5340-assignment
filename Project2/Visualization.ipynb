```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.style.use('seaborn')
sns.set(context='paper', style='darkgrid')
```

```python
df = pd.read_csv('varying_alpha.csv', index_col=0)
```

```python
def pairplot(alpha):
    sub_df = df.loc[df['alpha'] == alpha]
    g = sns.PairGrid(sub_df, vars=['g0', 'g1', 'g2', 'g3'])
    g = g.map_diag(sns.kdeplot)
    g = g.map_upper(plt.scatter)
    g = g.map_lower(sns.kdeplot, cmap="Blues_d")
    g.set(xlim=(-0.1, 1.1), ylim=(-0.1, 1.1))
    return g
```

```python
pairplot(0.01).savefig('report/images/alpha_0.01.png', facecolor='w')
```

```python
pairplot(0.1).savefig('report/images/alpha_0.1.png', facecolor='w')
```

```python
pairplot(1).savefig('report/images/alpha_1.png', facecolor='w')
```

```python
pairplot(10).savefig('report/images/alpha_10.png', facecolor='w')
```
